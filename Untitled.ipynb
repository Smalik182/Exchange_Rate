```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.style as style

# This enables Jupyter to display graphs
%matplotlib  inline

exchange_rate = pd.read_csv('euro-daily-hist_1999_2020.csv')
exchange_rate.head()
```

## Data Cleaning

We should rename some of the columns to have an easier time reading the data.
Note that the 'Period\Unit' which will be named 'Time' is also a object data type. We are going to convert it to a datetime data type.

```python
exchange_rate.rename(columns = {'[US dollar ]' : 'US_dollar', 'Period\\Unit:' : 'Time'},
                     inplace = True)
exchange_rate['Time'] = pd.to_datetime(exchange_rate['Time'])
exchange_rate.sort_values('Time', inplace = True)
exchange_rate.reset_index(drop = True, inplace = True)
```

Note that once we sorted the data set by the 'Time' column we then had to reset the index. 

```python
exchange_rate.head()
```

```python
euro_to_dollar = exchange_rate[['Time', 'US_dollar']].copy()
euro_to_dollar['US_dollar'].value_counts()
```

Note that there are 62 rows where the 'US_dollar' is '-', These rows should be dropped from the data set.
We should also convert 'US_dollar' to a float data type.

```python
euro_to_dollar = euro_to_dollar[euro_to_dollar['US_dollar'] != '-']
euro_to_dollar['US_dollar'] = euro_to_dollar['US_dollar'].astype(float)
euro_to_dollar.info()
```

## Rolling Mean

Now that the data is cleaned, lets visualize the evolution of the euro-dollar exchange rate. 

```python
plt.plot(euro_to_dollar['Time'],
         euro_to_dollar['US_dollar'])
plt.show()
```

We want a clearer view of the data. Since we do not need to see the changes happening daily we can opt to use the rolling mean of the exchange rate.

```python
plt.figure(figsize = (9,6))

plt.subplot(3,2,1)
plt.plot(euro_to_dollar['Time'], euro_to_dollar['US_dollar'])
plt.title('Original Values', weight = 'bold')

for i, rolling_mean in zip([2,3,4,5,6], # The remaining 6 subplots
                          [7,30,50,100,365]):   # The number of days the rolling mean will be
    plt.subplot(3,2,i)
    plt.plot(euro_to_dollar['Time'], euro_to_dollar['US_dollar'].rolling(rolling_mean).mean()) 
    plt.title('Rolling Window:' + str(rolling_mean), weight = 'bold')
    
plt.tight_layout()           # auto adjusts the padding between subplots
plt.show()
```

```python
euro_to_dollar['rolling_mean'] = euro_to_dollar['US_dollar'].rolling(30).mean()
euro_to_dollar
```

## How the Euro-Dollar has changed during times of crisis

We can take this data and visualize how the euro-dollar has changed during the 2007-2008 financial crisis and the coronovirus pandemic.

First we will make a graph to showcase the data during the financial crisis. We will focus on 2007-2008 particularly and we can use a line plot to showcase the comparision between 2006-2009.

```python
# Here we are copying the data into 2 seperate data sets

# from 2006 to 2009
financial_crisis = euro_to_dollar.copy()[
    (euro_to_dollar['Time'].dt.year >= 2006) &
    (euro_to_dollar['Time'].dt.year <= 2009)]

# from 2007 to 2008
financial_crisis_7_8 = euro_to_dollar.copy()[
    (euro_to_dollar['Time'].dt.year >= 2007) &
    (euro_to_dollar['Time'].dt.year <= 2008)]

financial_crisis_7_8
```

```python
# Adding the FiveThirtyEight style
style.use('fivethirtyeight')

### Adding the plot
fig,ax = plt.subplots(figsize=(8,3))
ax.plot(financial_crisis['Time'],
        financial_crisis['rolling_mean'],
        linewidth=1, color='blue')

### Highlighting the 2007-2008 period
ax.plot(financial_crisis_7_8['Time'],
        financial_crisis_7_8['rolling_mean'],
        linewidth=3, color='red')

# Changing the colar of tick labels
ax.tick_params(axis = 'x', colors = 'grey')
ax.tick_params(axis = 'y', colors = 'grey')

# Removing the labels on both axes
ax.set_xticklabels([])
ax.set_yticklabels([])

### Highlihting the peak of the crisis
ax.axvspan(xmin=13945.0, xmax=14125.0, ymin=0.09,  # trial and error used to find the exact peak location
           alpha=0.3, color='grey')

# To assign new labels we first need to find out the tic's cordinates
#ax.get_xticks()
# x tick output: array([13149., 13330., 13514., 13695., 13879., 14061., 14245., 14426., 14610.])
ax.get_yticks()
# y tick output: array([1.1, 1.2, 1.3, 1.4, 1.5, 1.6])

# assigning new labels
x = 13110.5
for year in ['2006', '2007', '2008', '2009']:
    ax.text(x, 1.13, year, alpha = 0.5, fontsize = 11)
    x += 365

y = 1.193
for rate in ['1.2', '1.3', '1.4', '1.5']:
    ax.text(12999.0, y, rate, alpha=0.5, fontsize=11)
    y += 0.1
    
# Adding the title and a subtitle
ax.text(12999.0, 1.67, "Euro-USD rate peaked at 1.59 during 2007-2008's financial crisis",
        weight='bold')
ax.text(12999.0, 1.63, 'Euro-USD exchange rates between 2006 and 2010',
        size=12)

# Adding a signarture
ax.text(12999.0, 1.07,' '* 120 + 'Source: European Central Bank',
        color = '#f0f0f0', backgroundcolor = '#4d4d4d',
        size=10)

# Adding some transparency to the grid
ax.grid(alpha = 0.5)

plt.show()
```

## Coronavirus Pandemic

Next lets make a visualization for the cornavirus pandemic. The pandemic first began in 2020 and is still on going with experts beleiving that the end of the pandemic is in sight but not here yet.

However the lastest entry in our data set is  '2021-01-08' so we will just stick with that.
For comparison, we will look at the data from 2016 to 2021. 

```python
# We create a new data set to hold the data for the pandemic period
corona_crisis_comparision = euro_to_dollar.copy()[
    (euro_to_dollar['Time'].dt.year >= 2016) &
    (euro_to_dollar['Time'].dt.year <= 2021)]


corona_crisis = euro_to_dollar.copy()[
    (euro_to_dollar['Time'].dt.year >= 2020) &
    (euro_to_dollar['Time'].dt.year <= 2021)]

# calculating the rolling mean
round(corona_crisis['rolling_mean'].min(),2)

corona_crisis
```

```python
style.use('fivethirtyeight')

# Adding the plot
fig,ax = plt.subplots(figsize=(8,3))
ax.plot(corona_crisis_comparision['Time'],
        corona_crisis_comparision['rolling_mean'],
        linewidth=1, color='blue')

# Adding the highlight for the pandemic duration
ax.plot(corona_crisis['Time'],
       corona_crisis['rolling_mean'], linewidth = 2.5, color = 'red')

# Changing the color of tick labels
ax.tick_params(axis ='x', colors = 'grey')
ax.tick_params(axis ='y', colors = 'grey')

# Removing labels on both axis
ax.set_xticklabels([])
ax.set_yticklabels([])

# To assign new labels we first need to find out the tic's cordinates
ax.get_xticks()
# x tick output: array([16436., 16801., 17167., 17532., 17897., 18262., 18628.])
ax.get_yticks()
# y tick output: array([1.  , 1.05, 1.1 , 1.15, 1.2 , 1.25])

# Assigning new labels

x = 16730
for year in  ['2016', '2017', '2018', '2019', '2020', '2021']:
    ax.text(x, 1.03, year, alpha=0.5, fontsize=15)
    x += 365
    
y = 1.05
for rate in  [1.05, 1.1 , 1.15, 1.2 , 1.25]:
    ax.text(16600, y, rate, alpha=0.5, fontsize=11)
    y += .05
    
# highlighting the period of pandemic
ax.axvspan(xmin=18265, xmax=18632, ymin=0.03,
           alpha=0.3, color='grey')

# adding title and labels
ax.text(16436,1.30,'EURO - USD rate change during 2016-2021 period of time',
        weight = 'bold', size=20, rotation = 0)
ax.text(16436,1.28,'Coronavirus Pandemic',
        weight = 'bold', size=15, rotation = 0)
ax.text(16436,1.15,'Rate',color ='black',alpha=0.3, size=15, rotation = 90)
ax.text(16436,1.01,'Year',color ='black',alpha=0.3, size=15, rotation = 0)


# adding the data source
ax.text(16420, 0.98,' '*70 + 'Source: European Central Bank',
        color = '#f0f0f0', backgroundcolor = '#4d4d4d',
        size=15)
plt.show()
```

## The Three US Presidencies

We can show comparatively how the euro-dollar changed under the last three presidents, President Bush (2001 - 2009), President Obama (2009 - 2017) and President Trump (2017 - 2020).

```python
presidents = euro_to_dollar.copy()[(euro_to_dollar['Time'].dt.year >= 2001) 
                                  & (euro_to_dollar['Time'].dt.year < 2021)]

bush = presidents.copy()[presidents['Time'].dt.year < 2009]

obama = presidents.copy()[(presidents['Time'].dt.year >= 2009)
                          & (presidents['Time'].dt.year < 2017)]

trump = presidents.copy()[(presidents['Time'].dt.year >= 2017)
                          & (presidents['Time'].dt.year < 2021)]   
```
